**Detecting outliers**

```python
from scipy import stats
# Import necessary libraries
import numpy as np

# 1. Z-Score Method
def detect_zscore_outliers(data, threshold=3):
    z_scores = stats.zscore(data)
    outliers = np.abs(z_scores) > threshold
    return outliers

# 2. IQR Method
def detect_iqr_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))
    return outliers

# Numerical columns for analysis
numerical_cols = ['Open', 'High', 'Low', 'Close', 'Volume']

# Create subplots for boxplots
# Visualization step using Matplotlib
plt.figure(figsize=(15, 5))
sns.boxplot(data=df[numerical_cols])
# Visualization step using Matplotlib
plt.title('Boxplot for Outlier Detection')
# Visualization step using Matplotlib
plt.xticks(rotation=45)
# Visualization step using Matplotlib
plt.show()

# Print outliers using both methods
print("\nOutlier Detection Results:")
for col in numerical_cols:
    print(f"\nAnalyzing {col}:")

    # Z-score outliers
    zscore_outliers = detect_zscore_outliers(df[col])
    print(f"Z-score outliers count: {zscore_outliers.sum()}")

    # IQR outliers
    iqr_outliers = detect_iqr_outliers(df[col])
    print(f"IQR outliers count: {iqr_outliers.sum()}")

    # Visualize outliers with scatter plot
# Visualization step using Matplotlib
    plt.figure(figsize=(10, 5))

# Visualization step using Matplotlib
    plt.subplot(1, 2, 1)
# Visualization step using Matplotlib
    plt.scatter(df.index[~zscore_outliers], df[col][~zscore_outliers], c='blue', label='Normal')
# Visualization step using Matplotlib
    plt.scatter(df.index[zscore_outliers], df[col][zscore_outliers], c='red', label='Outlier')
# Visualization step using Matplotlib
    plt.title(f'Z-score Outliers - {col}')
# Visualization step using Matplotlib
    plt.legend()

# Visualization step using Matplotlib
    plt.subplot(1, 2, 2)
# Visualization step using Matplotlib
    plt.scatter(df.index[~iqr_outliers], df[col][~iqr_outliers], c='blue', label='Normal')
# Visualization step using Matplotlib
    plt.scatter(df.index[iqr_outliers], df[col][iqr_outliers], c='red', label='Outlier')
# Visualization step using Matplotlib
    plt.title(f'IQR Outliers - {col}')
# Visualization step using Matplotlib
    plt.legend()

# Visualization step using Matplotlib
    plt.tight_layout()
# Visualization step using Matplotlib
    plt.show()

# Create a summary DataFrame with outlier information
outlier_summary = pd.DataFrame()
for col in numerical_cols:
    outlier_summary[f'{col}_zscore_outliers'] = detect_zscore_outliers(df[col])
    outlier_summary[f'{col}_iqr_outliers'] = detect_iqr_outliers(df[col])

# Display rows with outliers
print("\nRows containing outliers:")
outlier_rows = df[outlier_summary.any(axis=1)]
print(outlier_rows)
```

**Feature engineering**

```python
# Ensure the 'Date' column is a datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Sort the data by date (if not already sorted)
# Drop unnecessary columns from the dataset
df = df.sort_values(by='Date').reset_index(drop=True)

# Feature 1: Open-Close Spread
df['Open_Close_Spread'] = df['Open'] - df['Close']

# Feature 2: High-Low Spread
df['High_Low_Spread'] = df['High'] - df['Low']

# Feature 3: Quarter-End Flag
df['Quarter_End_Flag'] = df['Date'].dt.is_quarter_end.astype(int)

# Feature 4: Daily Returns
df['Daily_Return'] = df['Close'].pct_change()

# Feature 5: Moving Averages (e.g., 7-day, 14-day, and 30-day)
df['MA_7'] = df['Close'].rolling(window=7).mean()
df['MA_14'] = df['Close'].rolling(window=14).mean()
df['MA_30'] = df['Close'].rolling(window=30).mean()

# Feature 6: Price Momentum (Difference between Close price and 7-day Moving Average)
df['Momentum'] = df['Close'] - df['MA_7']

# Feature 7: Relative Strength Index (RSI)
# Calculate the daily gains and losses
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()

# Calculate RSI
rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))

# Feature 8: Exponential Moving Average (EMA) for trend detection
df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()

# Feature 9: MACD (Moving Average Convergence Divergence)
df['MACD'] = df['EMA_12'] - df['EMA_26']

# Feature 10: Target Variable
# If the next day's closing price is higher than the current day's, set the target to 1, otherwise 0
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)

# Drop rows with NaN values (caused by rolling calculations and target shift)
# Drop unnecessary columns from the dataset
df = df.dropna()

# Check the dataset after feature engineering
print(df.head())
```

**One-hot encoding**

```python
from sklearn.preprocessing import OneHotEncoder
# Import necessary libraries
import numpy as np

# Ensure the 'Date' column is a datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Sort the data by date
# Drop unnecessary columns from the dataset
df = df.sort_values(by='Date').reset_index(drop=True)

# Function to bin a numerical column into categories
def bin_column(column, bins, labels):
    return pd.cut(column, bins=bins, labels=labels, include_lowest=True)

# Define bins and labels
bin_labels = ['Low', 'Medium', 'High']
# Create bins dynamically using quantiles
for col in ['High', 'Low', 'Open', 'Close']:
    bins = np.quantile(df[col], [0, 0.33, 0.66, 1])  # Divide into 3 quantile-based bins
    df[f'{col}_Binned'] = bin_column(df[col], bins=bins, labels=bin_labels)

# One-Hot Encoding for the binned columns
# Drop unnecessary columns from the dataset
ohe = OneHotEncoder(sparse_output=False, drop='first')  # Use sparse_output instead of sparse
binned_cols = [f'{col}_Binned' for col in ['High', 'Low', 'Open', 'Close']]

# Apply one-hot encoding
encoded_features = ohe.fit_transform(df[binned_cols])
encoded_feature_names = ohe.get_feature_names_out(binned_cols)

# Add encoded features back to the original DataFrame
encoded_df = pd.DataFrame(encoded_features, columns=encoded_feature_names)
df = pd.concat([df, encoded_df], axis=1)

# Drop original binned columns (optional)
# Drop unnecessary columns from the dataset
df = df.drop(columns=binned_cols)

# Check the updated DataFrame
print(df.head())
```

```python
# Import necessary libraries
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

def objective(trial):
    # Suggest penalty and solver combinations
    penalty = trial.suggest_categorical('penalty', ['l1', 'l2', 'elasticnet', 'none'])
    solver = trial.suggest_categorical('solver', ['lbfgs', 'liblinear', 'saga'])

    # Ensure compatibility between penalty and solver
    if solver == 'lbfgs' and penalty not in ['l2']:
        raise optuna.exceptions.TrialPruned()
    if solver == 'liblinear' and penalty not in ['l1', 'l2']:
        raise optuna.exceptions.TrialPruned()
    if solver == 'saga' and penalty not in ['l1', 'l2', 'elasticnet']:
        raise optuna.exceptions.TrialPruned()
    if penalty == 'none' and solver != 'lbfgs':
        raise optuna.exceptions.TrialPruned()

    # Other hyperparameters
    C = trial.suggest_float('C', 0.01, 100.0, log=True)
    max_iter = trial.suggest_int('max_iter', 100, 500)

    # If elasticnet penalty, add l1_ratio
    if penalty == 'elasticnet':
        l1_ratio = trial.suggest_float('l1_ratio', 0.0, 1.0)  # Suggest value for l1_ratio

    # Configure the model
    model = LogisticRegression(penalty=None if penalty == 'none' else penalty,
                               C=C, solver=solver, max_iter=max_iter, random_state=42,
                               l1_ratio=l1_ratio if penalty == 'elasticnet' else None)

    # Calculate cross-validation score
    score = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc', error_score='raise').mean()
    return score

# Create and optimize the study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

print("Best Score:", study.best_value)
print("Best Parameters:", study.best_params)
```

```python
from hyperopt import hp, fmin, tpe, Trials
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
# Import necessary libraries
import numpy as np

# Define search space
search_space = {
    'C': hp.loguniform('C', np.log(0.01), np.log(100)),
    'solver': hp.choice('solver', ['lbfgs', 'liblinear', 'saga']),
    'max_iter': hp.choice('max_iter', [100, 200, 500]),
}

# Objective function
def objective(params):
    # Manually set penalty and l1_ratio based on solver
    solver = params['solver']

    if solver == 'liblinear':
        # liblinear supports only 'l1' or 'l2'
        penalty = np.random.choice(['l1', 'l2'])
        l1_ratio = None
    elif solver == 'lbfgs':
        # lbfgs supports only 'l2'
        penalty = 'l2'
        l1_ratio = None
    elif solver == 'saga':
        # saga supports 'l1', 'l2', or 'elasticnet'
        penalty = np.random.choice(['l1', 'l2', 'elasticnet'])
        if penalty == 'elasticnet':
            l1_ratio = np.random.uniform(0, 1)  # Randomly choose an l1_ratio between 0 and 1
        else:
            l1_ratio = None

    # Create LogisticRegression model with chosen parameters
    model = LogisticRegression(penalty=penalty, solver=solver, C=params['C'], max_iter=params['max_iter'], random_state=42, l1_ratio=l1_ratio)
    score = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc').mean()

    return -score

# Perform optimization
trials = Trials()
best = fmin(fn=objective, space=search_space, algo=tpe.suggest, max_evals=50, trials=trials)

print("Best Parameters:", best)
```
